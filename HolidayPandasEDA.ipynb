```python
import pandas as pd
import numpy as np
import json
from functools import reduce
```

```python
with open('./new_traits.json') as fp:
    data = json.load(fp)
    
df = pd.json_normalize(data)
df.sample(10)
```

```python
import itertools

holiday_body = list(set([x['newTraits']['BODY'] for x in data]))
holiday_head = list(set([x['newTraits']['HEAD'] for x in data]))

new_trait_combos = [{'BODY': x[0], 'HEAD': x[1]} for x in itertools.product(holiday_body, holiday_head)]
new_trait_combos
```

```python
df_changed = df.loc[(df['newTraits.HEAD'] != 'Original') | (df['newTraits.BODY'] != 'Original')]
df_unchanged = df.loc[~df.index.isin(df_changed.index)]

print(f'total pandas: {len(df)}')
print(f'changed pandas: {len(df_changed)}')
print(f'unchanged pandas: {len(df_unchanged)}')
```

```python
df['traits.❤️'].value_counts()
```

```python
def print_trait_stats(trait):
    trait = f'traits.{trait}'
    print(df_changed[trait].value_counts()/df[trait].value_counts())
    
print_trait_stats('❤️')
```

```python
df_ = df.copy(deep=True)
df_['traits.BODY'] = df_.apply(
    lambda x: x['newTraits.BODY'] if x['newTraits.BODY'] != 'Original' else x['traits.BODY'], 
    axis=1
)
df_['traits.HEAD'] = df_.apply(
    lambda x: x['newTraits.HEAD'] if x['newTraits.HEAD'] != 'Original' else x['traits.HEAD'], 
    axis=1
)
```

```python
def filter_by_traits(df, **kwargs):
    df_ = df
    filters = [
        df_[f'newTraits.{k}'] == f'{v}'
        for k, v in kwargs.items()
    ]
    return df_[reduce(np.logical_and, filters)]

def get_pct_by_traits(df, **kwargs):
    df_filtered = filter_by_traits(df, **kwargs)
    return len(df_filtered)/len(df_)
```

```python
trait_pct = []
for trait in new_trait_combos:
    trait['% of pandas'] = get_pct_by_traits(df, BODY=trait['BODY'], HEAD=trait['HEAD'])
    trait_pct.append(trait)
    
df_new_trait_pct = pd.DataFrame(trait_pct)
df_new_trait_pct = df_new_trait_pct.sort_values('% of pandas').reset_index(drop=True)
df_new_trait_pct
```
